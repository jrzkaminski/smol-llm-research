```python
import pandas as pd
from termo import Termo
```

```python
with open("../gnn4taskplan_modeling/tmdb/tool.json", "r") as file:
    text = file.read()
text
```

```python
model = "meta-llama/llama-3.3-70b-instruct"
model_params = {
    "temperature": 0.0,
}
```

```python
termo = Termo(text)
terms = termo.extract_terms(model=model, **model_params)
df = pd.DataFrame(terms, columns=["Term", "Start Index", "End Index", "Sentence Index"])
df
```

```python
termo["terms"] = terms

acronyms = termo.extract_acronyms(model=model, max_length_split=2000, **model_params)
df = pd.DataFrame(list(acronyms.items()), columns=["Acronym", "Term"])
df
```

```python
termo["terms"] = terms

definitions = termo.extract_definitions(
    model=model, max_length_split=2000, **model_params
)
df = pd.DataFrame(list(definitions.items()), columns=["Term", "Definition"])
df
```

```python
termo["terms"] = terms

relationships = termo.extract_relationships(
    model=model, max_length_split=2000, **model_params
)
df = pd.DataFrame(relationships, columns=["Term1 ->", "Relationship ->", "Term2"])
df
```
